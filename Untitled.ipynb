```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import warnings
warnings.filterwarnings('ignore')
```

Datan är taget från UCl machine learnings repository. 8 stycken features enligt nedan. med Pandas read csv så kan vi hantera datan då det lagas i numpy arrayer.

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data'
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv(url, header=None, names=col_names)
```

För att inte bygga upp komplexiteten av din algoritm så vill du ha så få features som möjligt (enligt 'the curse of dimensionality'). Du vill med så få features som möjligt få ut så mycket som möjligt. Därmed så kollar vi efter växelverkan mellan alla features och om vi ser någon feature som har en hög växelverkan (ju rödare ruta som möjligt i heatmappen nedan) så kan denna tas bort då den lilla nya informationen den ger gör att den inte är värd att ta med. Delar av datan sparas i matrisen X, samt responsvektorn Y.

```python
X = pima.iloc[:,0:8]
y = pima.iloc[:,-1]
```

Vi splittrar upp vår data i 5 bitar. 20% blir vår test data och 80% vår träningsdata.

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Eftersom det är ett klassifikationsproblem så använder jag ett par modeller för att hitta den bästa.

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

En modellista skapas.

```python
models = []
models.append(("LR",LogisticRegression()))
models.append(("NB",GaussianNB()))
models.append(("KNN",KNeighborsClassifier()))
models.append(("DT",DecisionTreeClassifier()))
models.append(("SVM",SVC()))
```

Cross validation används och den bästa modellen för detta experimentet skrivs ut. Den bästa modellen är bestämd genom att ge viken noggrannhet den ger. Det fanns exempel då man använde andra metoder med för att hitta för hur the standard deviation samt variansen berode, men detta är mitt första exempel så kanske senare så får jag gräva vidare. Från detta ser vi i alla fall att logistic regression (LR) är den bästa modellen för detta sammanhanget.

```python
results = []
names = []
for name,model in models:
    kfold = KFold(n_splits=10, random_state=22)
    cv_result = cross_val_score(model,X_train,y_train, cv = kfold,scoring = "accuracy")
    names.append(name)
    results.append(cv_result)
for i in range(len(names)):
    print(names[i],results[i].mean())
```

```python
LR = LogisticRegression()
LR.fit(X_train,y_train)
predictions = LR.predict(X_test)
```

För att se hur noggrannt vårt estimat är så använder vi accuracy score.

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,predictions))
```

Fast detta resultatet är inte hela sanningen. Detta tar inte till hänsyn hur datan distribuerades eller vilka slags fel vi får.

```python
conf = metrics.confusion_matrix(y_test,predictions)
label = ["0","1"]
sns.heatmap(conf, annot=True, xticklabels=label, yticklabels=label)
```

Denna matris visar hur algoritmen klarade av fallen. 98 gånger fick vi en false-negative (algoritmen bestämde att patienten inte har diabetes korrekt). 9 gånger fick vi en true-negative (algoritmen bestämde att patienten har diabetes inkorrekt Type 1 error). 19 gånger fick vi en false-positive (algoritmen bestämde att patienten inte har diabetes inkorrekt Type 2 error).
Och slutligen 28 gånger av fallen så fick en true-positive (algoritmen bestämde att patienten har diabetes korrekt). Så även om noggrannheten är väldigt hög så betyder det inte att resultatet är det bästa som du vill ha. Det kan bero på din träningsdata. Om majoriteten av datan är samplad av en viss typ så kommer den att vara biased om en liknande data ska bestämmas. Därmed så är det alltid bra att kolla på confusion matrixen som visar exakt hur resultatet ser ut.

Beroende på vad vi är intresserade av så kan vi ändra toleransen på klassifikationen. För att hitta en bra tolerans kan man forma en ROC kurva som  visar True-positive rate mot False-positive rate. Så beroende på hur datan/problemet ser ut så kan man hitta punkter på kurvan och se hur din algoritm verkligen beter sig.

```python
fpr, tpr, thresholds = metrics.roc_curve(y_test, predictions)
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for diabetes classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
```

```python
print(metrics.roc_auc_score(y_test, predictions))
```

Med denna siffra så kan man få en uppfattning om hur bra klassifieraren är, detta är arean under ROC kurvan.
